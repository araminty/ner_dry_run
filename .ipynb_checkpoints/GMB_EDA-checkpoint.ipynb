```python
import os

 # naming the columns according to specifications in the readme
column_names = ['token', 'penn', 'lemma', 'ner', 'wordnet', 'verbnet', 'semantic', 'animacy', 'supertag', 'boxer']
df = pd.DataFrame(columns = column_names)


dirName = "../gmb-2.2.0/data/"

listOfFiles = list()
for (dirpath, dirnames, filenames) in os.walk(dirName):
    for file in filenames:
        if ".tags" in file:
            listOfFiles += [os.path.join(dirpath, file)]

```

```python
from tqdm.notebook import tqdm

for path in tqdm(listOfFiles[:100]):
    df_read = pd.read_csv(path, sep='\t', names=column_names)
    df = pd.concat([df, df_read])
```

```python
print(df[df['ner']!='O']['ner'].value_counts().sum())
df[df['ner']!='O']['ner'].value_counts()
```

```python
df.head()
```

One thing that worries me is the possibility that we could overfit on the tokens in question.  While we dont want to blind ourselves to these tokens, generalizability seems important given the context of information extraction.

```python
ners = df[df['ner']!='O']['token'].value_counts()
not_ners = df[df['ner']=='O']['token'].value_counts()
```

```python
ners[ners>20]
```

```python
not_ners[not_ners.index.isin(ners.index)].head(20)
```

This do not seem like named entities...

```python
top5overlaps = not_ners[not_ners.index.isin(ners.index)].head(5).index

df[(df['token'].isin(top5overlaps)) & (df['ner']!='O')]
```

The labeling schema is taken several sequential tokens as the ner, some of which are annoyingly generic.  So they will need to be combined purely by the positions.  This does largely alleviate my fears about overfitting however because the matches will be several consecutive words.
